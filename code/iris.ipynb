```python
from datasets import load_dataset

dataset = load_dataset("scikit-learn/iris")
print(dataset)
```

```python
dataset["train"][0]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# استخراج بخش train به صورت DataFrame
df = pd.DataFrame(dataset["train"])

# نمایش چند سطر اول داده
print(df[10:12])
```

```python
# بررسی توزیع کلاس‌ها
print("\nClass distribution:")
print(df['Species'].value_counts())
```

```python
# پلات جفتی از ویژگی‌ها با رنگ بر اساس نوع گل
sns.pairplot(df, hue="Species", diag_kind="kde")
plt.suptitle("Iris Dataset Feature Relationships", y=1.02)
plt.show()
```

```python
# پلات توزیع هر ویژگی برای درک بهتر تفاوت بین کلاس‌ها
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
plt.figure(figsize=(10,6))
for f in features:
    sns.kdeplot(data=df, x=f, hue="Species", fill=True, alpha=0.4)
plt.title("Feature Distributions by Species")
plt.show()
```
